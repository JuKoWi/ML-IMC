```julia
"""
function readloss(outname)

Reads loss data from ML-IMC output file
"""
function readloss(outname)
    lossdata = []
    file = open(outname, "r")
    lines = readlines(file)
    for line in lines
        if length(split(line)) > 0 && split(line)[1] == "Loss"
            append!(lossdata, parse(Float64, split(line)[end]))
        end
    end
    close(file)
    iters = 1:length(lossdata)
    return(iters, lossdata)
end;
```

```julia
"""
histname(rdfname)

Reads distance histogram file produced
by ML-IMC.jl
"""
function readhist(histname)
    file = open(histname, "r")
    println("Reading histogram data from $(histname)...")
    lines = readlines(file)
    ncomments = 1
    nlines = length(lines) - ncomments
    bins = zeros(nlines)
    hist = zeros(nlines)
    for i in (1 + ncomments):length(lines)
        rdfline = split(lines[i])
        if length(rdfline) == 2
            bins[i - ncomments] = parse(Float64, rdfline[1])
            hist[i - ncomments] = parse(Float64, rdfline[2])
        end
    end
    return(bins, hist)
    close(file)
end;
```

```julia
function readenergies(outname)
    steps = []
    energies = []
    file = open(outname, "r")
    println("Reading energy data from $(outname)...")
    lines = readlines(file)
    for line in lines
        splittedLine = split(line)
        if length(splittedLine) > 0
            if splittedLine[1] == "#" && splittedLine[2] == "Step"
                append!(steps, parse(Int, splittedLine[4]))
            elseif splittedLine[1] != "#"
                append!(energies, parse(Float64, splittedLine[1]))
            end
        end
    end
    return(steps, energies)
end;
```

```julia
itersRDF1, lossdataRDF1 = readloss("ML-IMC.out");
```

```julia
plt = plot(itersRDF1, lossdataRDF1, title="Training with RDF data", 
    label="Loss" * L": (RDF^{NN} - RDF^{ref})^2", marker=(:dot), color=palette(:tab10)[1])
xlabel!("Iteration")
ylabel!("Loss")
display(plt)
savefig("Loss.pdf")
```

```julia
binsRDF, rdfNN1 = readhist("rdfNN-iter-01.dat");
binsRDF, rdfNN5 = readhist("rdfNN-iter-05.dat");
binsRDF, rdfNN25 = readhist("rdfNN-iter-25.dat");
```

```julia
plt = plot(bins, rdfref, xlim = (2, 15), lw=3, 
    labels="Reference", title="RDF", legend=:best, color=palette(:tab10)[1])
plot!(binsRDF, rdfNN1, lw=2, 
    linestyle=:dash, label="Initial guess", color=palette(:tab10)[2])
plot!(binsRDF, rdfNN5, lw=2, 
    linestyle=:dash, label="Iteration 5", color=palette(:tab10)[3])
plot!(binsRDF, rdfNN25, lw=2, 
    linestyle=:dash, label="Iteration 25", color=palette(:tab10)[4])
xlabel!("Distance, Å")
ylabel!("g(r)")
display(plt)
savefig("training.pdf")
```

At iteration 25 the convergence is almost perfect!
