```julia
"""
function readloss(outname)

Reads loss data from ML-IMC output file
"""
function readloss(outname, keyword)
    lossdata = []
    file = open(outname, "r")
    lines = readlines(file)
    for line in lines
        if length(split(line)) > 0 && split(line)[1] == keyword
            append!(lossdata, parse(Float64, split(line)[end]))
        end
    end
    close(file)
    iters = 1:length(lossdata)
    return(iters, lossdata)
end;
```

```julia
strLoss = readloss("mc.out", "Descriptor");
```

```julia
regLoss = readloss("mc.out", "Regularization");
```

```julia
plot(strLoss[1], strLoss[2], lw=2, label="RDF loss")
plot!(regLoss[1], regLoss[2], lw=2, label="Regularization loss")
plot!(strLoss[1], strLoss[2] .+ regLoss[2], lw=2, label="Total loss")
plot!(legend=:outertopright, xlim=(0, 20), ylim=(0, 100))
xlabel!("Iteration")
ylabel!("Loss")
```
